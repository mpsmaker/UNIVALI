```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pylab
```

```python
print("comeco:" + str(-np.pi))
print("fim:" + str(np.pi))
print("intervalo:" + str(np.pi/100))

w = np.arange(-np.pi, np.pi, np.pi/100) #W vai de menos pi atÃ© pi sobre 100 

Num = 2*np.cos(w) + 1 # Numerador
Den = 1  # - a*np.exp(-1j*w) Denominador
X = Num/Den #x = numerador/denominador

print("X:" + str(X))
print("X.shape:" + str(X.shape))

Mod_X = abs(X) #mÃ³dulo corresponde ao mÃ³dulo de X
print("Mod")
print(Mod_X)
Fase_X = np.angle(X) #Fase corresponde ao Ã¢ngulo
print("Fase")
print(Fase_X.shape)
print(Fase_X)
```

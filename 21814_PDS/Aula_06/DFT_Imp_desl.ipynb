### Plota a DFT de impulsos deslocados
#### x[n] = delta[n+1] + delta[n] + delta[n-1] 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 100
w = np.arange(-np.pi, np.pi, np.pi/Fs)

X = 1 + 2 * np.cos(w)
```

```python

#Mod_X = np.abs(X)
#Fase_X = np.angle(X)

plt.grid()
plt.plot(w,X)
#plt.plot(w,Fase_X)

plt.show()
```

```python
# Plotando em HZ
#Fs = 1000
f_hz = (w/np.pi)*(Fs/2)

plt.grid()
plt.plot(f_hz,X)

plt.show()
```

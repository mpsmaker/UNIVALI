## Plotando o modulo e fase da TFD
#### do sinal a^n.u[n]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

a = .5

Fs = 1000

w = np.arange(-2*np.pi, 2*np.pi, np.pi/Fs)

Num = 1
Den = 1 - a * np.exp(-1j*w)
X = Num / Den

Mod_X = np.abs(X)
Fase_X = np.angle(X)
```

```python
plt.grid()
plt.plot(w,Mod_X)
plt.plot(w,Fase_X)

plt.show()
```

```python
# Plotando em HZ
#Fs = 1000
f_hz = (w/np.pi)*(Fs/2)

plt.grid()
plt.plot(f_hz,Mod_X)
plt.plot(f_hz,Fase_X)

plt.show()
```

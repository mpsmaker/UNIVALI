```python
import numpy as np
import matplotlib.pyplot as plt

# D(Z) = 1 + 1.5Z^-1 + 2Z^-2
#        |    |        |
#        a    b        c
#--------------------------- variaveis #[1, 1.5, 2]
#var_na = 1
#var_nb = 1.5
#var_nc = 2
#-------------
#var_da = 1 
#var_db = 0 
#var_dc = 0 

#--------------------------- variaveis 5b #[0.2, 0.3, 0.3, 0.2] / [1, 0, 0, 0]
#var_na = 0.2
#var_nb = 0.3
#var_nc = 0.3
#var_nd = 0.2
#-------------
#var_da = 1 
#var_db = 0 
#var_dc = 0
#var_dd = 0

#--------------------------- variaveis 6a #[-0.2, -0.4, -0.4, 1] / [0.6, 1, 0, 0]
#var_na = -0.2
#var_nb = -0.4
#var_nc = -0.4
#var_nd = 1
#-------------
#var_da = 0.6 
#var_db = 1 
#var_dc = 0 
#var_dd = 0 

#--------------------------- variaveis 6b #[1, 1, 0.41] / [0, 1, 0.8]
var_na = 1
var_nb = 1
var_nc = 0.41
#-------------
var_da = 0 
var_db = 1 
var_dc = 0.8 

N = [var_na, var_nb, var_nc] #, var_nd]
D = [var_da, var_db, var_dc] #, var_dd] 

Num = N
Den = D 

# Set up vector for zeros
z = np.roots(Num)

# Set up vector for poles
p = np.roots(Den)

fig, ax = plt.subplots(figsize=(6, 6))  # Define o tamanho da figura

zplane = plt.scatter(np.real(z), np.imag(z), marker='o', color='b', label='Zeros')
zplane = plt.scatter(np.real(p), np.imag(p), marker='x', color='r', label='Poles')

# Adicionar rtulos aos pontos Z (zeros)
for zi in z:
    ax.annotate(f'Z({zi.real:.2f}, {zi.imag:.2f})', (zi.real, zi.imag), textcoords="offset points", xytext=(-15,-10), ha='center')

# Adicionar rtulos aos pontos P (polos)
for pi in p:
    ax.annotate(f'P({pi.real:.2f}, {pi.imag:.2f})', (pi.real, pi.imag), textcoords="offset points", xytext=(15,5), ha='center')

# Adicionar um crculo de raio 1
theta = np.linspace(0, 2 * np.pi, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)
plt.plot(x_circle, y_circle, linestyle='--', color='g', label='Crculo (Raio 1)')

# Definir limites dos eixos para que o grfico fique simtrico
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])

plt.axvline(0, color='k', linestyle='--', linewidth=1)
plt.axhline(0, color='k', linestyle='--', linewidth=1)
plt.grid()
plt.legend()
plt.title('Pole/Zero Plot Example')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```

# -----------------------------------------------------------------------------------------------------
# Detector de Potencia do Sinal
# Calcular uma Potencia de n amostras definidos para um frame de 80 - 160 amostras
# comparar com o Th também definido e caso a potencia for maior que o Th, sinalizar com um
# Impulso unitário para cada Pot[n] > Th, senão 0
# -----------------------------------------------------------------------------------------------------

```python
# Importando bibliotecas:
# Nesta seção, importamos as bibliotecas numpy e matplotlib.pyplot. 
# A biblioteca numpy é usada para cálculos numéricos eficientes e manipulação de arrays, 
# enquanto matplotlib.pyplot é usada para plotar gráficos.

import numpy as np
import matplotlib.pyplot as plt


# Definindo funções para os sinais discretos básicos:
# Aqui, definimos quatro funções que representam os sinais discretos básicos. 
# impulso_unitario retorna um impulso unitário em uma posição específica. 
# degrau_unitario retorna um degrau unitário a partir de uma posição específica. 
# sinal_sinusoidal retorna um sinal sinusoidal com amplitude, frequência angular e fase especificadas. 
# sinal_exponencial retorna um sinal exponencial com amplitude e fator de crescimento especificados.

def impulso_unitario(n, n0=0):
    return np.where(n == n0, 1, 0)

def degrau_unitario(n, n0=0):
    return np.where(n >= n0, 1, 0)

def sinal_sinusoidal(n, A=1, omega=2*np.pi, phi=0):
    return A * np.sin(omega * n + phi)

def sinal_exponencial(n, A=1, alpha=0.5):
    return A * np.exp(alpha * n)


# Definindo parâmetros e gerando exemplos de sinais:
# Nesta parte, definimos um array n que contém amostras no intervalo de -10 a 10. 
# Em seguida, geramos exemplos dos sinais discretos básicos para este intervalo, 
# usando as funções que definimos anteriormente.

n = np.arange(-10, 11)  # Amostras no intervalo de -10 a 10

# Exemplos das funções

impulso = impulso_unitario(n, n0=0)
degrau = degrau_unitario(n, n0=0)
sinusoidal = sinal_sinusoidal(n, A=1, omega=np.pi/4)
exponencial = sinal_exponencial(n, A=1, alpha=-0.1)


# Plotagem dos sinais
# Aqui, criamos uma figura de tamanho 10x8 polegadas para a plotagem. 
# Usamos a função plt.subplot para criar uma matriz de subplots 2x2, onde cada subplot é usado para um sinal. 
# Para cada subplot, usamos plt.stem para criar um gráfico de haste (stem plot) dos sinais em relação a n. 
# Definimos títulos, rótulos dos eixos e ajustamos o layout dos subplots com plt.tight_layout(). 
# Finalmente, usamos plt.show() para exibir os gráficos.

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.stem(n, impulso)
plt.title('Impulso Unitário')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.stem(n, degrau)
plt.title('Degrau Unitário')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.stem(n, sinusoidal)
plt.title('Sinal Sinusoidal')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 4)
plt.stem(n, exponencial)
plt.title('Sinal Exponencial')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

# Amostrando um Sinal Continuo no tempo
#
# x(t) = A.Sin (w0 t)  w0 = 2.pif0   f0 = 100hz    t0 = 1 Divide f0 = 10ms
# x(t) = Sin (2.pi.100 t)
#
# Amostras
#
# x(t) = A. Sin (2.pi.f0 n Ts)
# 
# x[n] =   Sin (2. pi f0 n) 
#                    --
#                    Fs
#
# n   x[n]
# -----------------------
# 0   0
# 1   sin(2.pi)
#         ----
#         10
# 2
# 3
# 4
# 6
# 7
# 8
# 9
# 10  Sin(2.pi)
#
# Telecomunicacoes
#
# Fs = 8KHz
# ts = 125 ms

```python
# Importando bibliotecas:
# Nesta seção, importamos as bibliotecas numpy e matplotlib.pyplot. 
# A biblioteca numpy é usada para cálculos numéricos eficientes e manipulação de arrays, 
# enquanto matplotlib.pyplot é usada para plotar gráficos.

import numpy as np
import matplotlib.pyplot as plt


# Criando o sinal analógico e plotando:
# Nesta parte, definimos o intervalo de tempo t usando np.arange para criar um vetor 
# que varia de -5e-3 a 5e-3 em incrementos de Dt (5e-5). 
# Em seguida, criamos o sinal analógico xa utilizando a função np.cos para 
# gerar um sinal cosseno com frequência f. 
# Plotamos o sinal analógico usando plt.plot, configuramos os rótulos dos eixos 
# e o título do gráfico, e ativamos a grade com plt.grid(True).

Dt = 5e-5
t = np.arange(-5e-3, 5e-3 + Dt, Dt)

# Plota o sinal analógico
f = 1000
xa = np.cos(2 * np.pi * f * t)

plt.subplot(2, 1, 1)
plt.plot(t * 1000, xa)
plt.xlabel('Tempo [ms]')
plt.ylabel('xa(t)')
plt.title(f'Sinal de Freq = {f} Hz')
plt.grid(True)


# Criando o sinal discreto e plotando:
# Aqui, definimos a taxa de amostragem Fs e o período de amostragem Ts. 
# Criamos um vetor n que varia de -20 a 20, representando as amostras discretas. 
# Calculamos o sinal discreto xd usando a fórmula do cosseno discreto. 
# Usamos plt.stem para criar um gráfico de haste (stem plot) do sinal discreto, 
# configuramos a cor para vermelho.

Fs = 8000
Ts = 1 / Fs
f = 400
n = np.arange(-20, 21)
xd = np.cos(2 * np.pi * n * f / Fs)
N = len(n)

plt.stem(n * Ts * f, xd, 'r')


# Plotando o sinal amostrado:
# Aqui, criamos um novo subplot para o sinal amostrado e usamos plt.stem novamente para criar um gráfico de haste. 
# Definimos títulos, rótulos dos eixos e ativamos a grade. Finalmente, usamos plt.tight_layout() para ajustar 
# o layout dos subplots e plt.show() para exibir os gráficos.

plt.subplot(2, 1, 2)
plt.stem(n, xd)
plt.title('Sinal amostrado')
plt.xlabel('Amostras')
plt.ylabel('x[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## **Sinal Analogico**

```python
t = np.arange(-5e-3,5e-3,5e-5)
f = 1000
xa_cos = 2*math.pi*f*t
xa = []
for i in xa_cos:
  xa.append(math.cos(i))
# xa = math.cos(2*math.pi*f*t)
# Trocar plot por stem pra gerar um sinal correto
# np.stem()
plt.xlabel("Tempo [ms]")
plt.ylabel("xa(t)")
plt.title(f"Sinal analogico - Sinal de Freq {f} Hz")
plt.plot(t*1000, xa) 
plt.show() 
```

## **Sinal Discreto**

```python
Fs = 8000
Ts = 1/Fs
n = np.arange(-20,20,1)
N = len(n)
xd_cos = 2*math.pi*n*f/Fs
xd = []
for i in xd_cos:
  xd.append(math.cos(i))
plt.title("Sinal discreto")
plt.stem(n*Ts*1000, xd, use_line_collection = True)
```

## **Sinal Amostrado**

```python
plt.xlabel("Amostras")
plt.ylabel("x[n]")
plt.title("Sinal amostrado")
plt.stem(n, xd, use_line_collection = True)
```

## **Sequencia Sinusoidal**

```python
Fs = 8000 # Frq d/ amostragem
f0 = 100 # Frq
t = 100 # tempo de exec ms
A = 1000 # valor Amplitude
t = np.arange(t)
y = np.cos(2 * np.pi * f0 * t / Fs)
y = A * y
plt.stem(t, y)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

```python
with open('/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/seq_sinusoidal.pcm', 'wb') as f:
    np.array(y, dtype=np.int16).tofile(f)
f.close()
```

# **Sequencia Exponencial**

```python
A = 10000 #amp
tam = 10 # tam
n = np.arange(0, tam, 1)
a1 = 2 #0.5 2 e -0.5

yn = []
for i in range(len(n)):
  yn.append(A*(a1**(i)))
plt.stem(n,yn)
plt.show()
```

```python
with open('/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/seq_exponencial.pcm', 'wb') as f:
    np.array(yn, dtype=np.int16).tofile(f)
f.close()
```

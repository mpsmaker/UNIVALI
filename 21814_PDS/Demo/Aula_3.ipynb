# **Media Movel**

```python
dado = np.memmap("/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/impulso.pcm", dtype='h', mode='r')
# dado = np.memmap("/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/sweep.pcm", dtype='h', mode='r')

# print("Entrada")
# count = 0
# for x in range(len(dado)):
#   if dado[x] > 0:
#     print("Pos: " + str(x) + " - Valor: " + str(dado[x]))
#     count = count + 1
# print("Valores totais >0 : " + str(count))
 
k = int(input("Digite o tamanho de mdia (4, 8, 16, 32, 64):"))

ret = np.cumsum(dado, dtype='int16') 
ret[k:] = ret[k:] - ret[:-k] 
saida = ret[k - 1:] / k

plt.figure(figsize=[10, 5])
#Plota o original
plt.subplot(211)
plt.title("Entrada")
plt.grid(True)
plt.stem(dado)

#Plota a media
plt.subplot(212)
plt.title("Saida")
plt.grid(True)
plt.stem(saida)
plt.tight_layout()

plt.show()

# print("\nSaida")
# count = 0
# for x in range(len(saida)):
#   if saida[x] > 0:
#     print("Pos: " + str(x) + " - Valor: " + str(saida[x]))
#     count = count + 1
# print("Valores totais >0 : " + str(count))

#Plota com zoom
saida_z = []
x = int(input("Digite o tamanho do zoom (10, 20, 50, 80...):\n"))
for y in range(len(saida)):
  if y >= x and y <= len(saida)-x:
    saida_z.append(saida[y])
plt.subplot(212)
plt.title("Saida")
plt.grid(True)
plt.stem(saida_z)
plt.tight_layout()

plt.show()

with open('/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/media_movel.pcm', 'wb') as f:
    np.array(saida, dtype=np.int16).tofile(f)
f.close()
```

# **Delay**

```python
# dado = np.memmap("/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/impulso.pcm", dtype='h', mode='r')
dado = np.memmap("/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/alo.pcm", dtype='h', mode='r')

Fs = 8000 # Frq de amostragem
t1 = 300*10**-3 # tempo 1 para o n1
t2 = 500*10**-3 # tempo 2 para o n2
a0 = .5 # amostra 0
a1 = .3 # amostra 1
a2 = .2 # amostra 2
n1 = int(t1*Fs) # 
n2 = int(t2*Fs)

print("n1: " + str(n1))
print("n2: " + str(n2))

# Fs = int(input("Frequncia de amostragem:"))
# t1 = float(input("T1:"))
# t2 = float(input("T2:"))
# a0 = float(input("a0:"))
# a1 = float(input("a1:"))
# a2 = float(input("a2:"))
# n1 = int(t1 * Fs)
# n2 = int(t2 * Fs)

delay = np.zeros(n2)
saida = np.zeros(len(dado))

for pointer in range(len(dado)):
  delay[0] = dado[pointer]
  y = a0 * delay[0] + a1 * delay[n1-1] + a2 * delay[n2-1]
  for i in range(n2):
    delay[n2-i-1] = delay[n2-i-2]
    saida[pointer] = y

plt.subplot(211)
pylab.stem(dado)
plt.subplot(212)
pylab.stem(saida)
pylab.show()

print("\nEntrada:")

count = 0
for x in range(len(dado)):
  if dado[x] > 0:
    print("Pos: " + str(x) + " - Valor: " + str(dado[x]))
    count = count + 1
print("Valores totais >0 : " + str(count))

print("\nSaida")

count = 0
for x in range(len(saida)):
  if saida[x] > 0:
    print("Pos: " + str(x) + " - Valor: " + str(saida[x]))
    count = count + 1
print("Valores totais >0 : " + str(count))

#Plota com zoom
saida_z = []
x = int(input("Digite o tamanho do zoom (10, 20, 50, 80...):\n"))
for y in range(len(saida)):
  if y >= 100 and y <= len(saida)-x:
    saida_z.append(saida[y])
plt.subplot(212)
plt.title("Saida")
plt.grid(True)
plt.stem(saida_z)
plt.tight_layout()

with open('/content/gdrive/MyDrive/PDS - Processamento Digital de Sinais/delay_alo.pcm', 'wb') as fid:
    np.array(saida, dtype=np.int16).tofile(fid)
fid.close()
```

# **Eco**

```python
print("comeco:" + str(-np.pi))
print("fim:" + str(np.pi))
print("intervalo:" + str(np.pi/100))

w = np.arange(-np.pi, np.pi, np.pi/100) #W vai de menos pi até pi sobre 100 

Num = 2*np.cos(w) + 1 # Numerador
Den = 1  # - a*np.exp(-1j*w) Denominador
X = Num/Den #x = numerador/denominador

print("X:" + str(X))
print("X.shape:" + str(X.shape))

Mod_X = abs(X) #módulo corresponde ao módulo de X
print("Mod")
print(Mod_X)
Fase_X = np.angle(X) #Fase corresponde ao ângulo
print("Fase")
print(Fase_X.shape)
print(Fase_X)

#Plota o módulo/Magnitude
plt.subplot(2,1,1)
plt.plot(w, Mod_X,'r')
plt.title('Magnitude de X[n]')

#Plota a fase
plt.subplot(2,1,2)
plt.plot(w, Fase_X)
plt.title('Fase de X[n]')


plt.tight_layout()
plt.show()

print("Como no exemplo PDF")
w = np.arange(-np.pi, np.pi, np.pi/100) #W vai de menos pi até pi sobre 100 
a = 0.5
Num = 1 # Numerador
Den = 1 - a*np.exp(-1j*w) # Denominador
X = Num/Den #x = numerador/denominador

Mod_X = abs(X) #módulo corresponde ao módulo de X
Fase_X = np.angle(X) #Fase corresponde ao ângulo

#Plota o módulo/Magnitude
plt.subplot(2,1,1)
plt.plot(w, Mod_X,'r')
plt.title('Magnitude de X[n]')

#Plota a fase
plt.subplot(2,1,2)
plt.plot(w, Fase_X)
plt.title('Fase de X[n]')


plt.tight_layout()
plt.show()
```
